```python
from scipy.cluster.hierarchy import dendrogram
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20, 16))
plt.title('Dendrogram')
plt.xlabel('Index Numbers')
plt.ylabel('Distance')
dendrogram(
    hcluster,
    leaf_rotation=90,
    leaf_font_size=8.,
    
)
# plt.ylim((0, 2))

plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
max_dist=10
clusters=fcluster(hcluster, max_dist, criterion='distance')
print set(clusters)
clusters
```

```python
plt.scatter(data[:, 0], data[:, 2], c=clusters, cmap='viridis')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
wines=pd.read_csv('../datasets/wine.csv')
```

```python
quality=wines.quality
```

```python
wines.drop('quality', axis=1, inplace=True)
```

```python
wines.shape
```

```python
wines_standardized=StandardScaler().fit_transform(wines)
```

```python
wines_standardized.shape
```

```python
wines_cluster=linkage(wines_standardized, 'ward')
```

```python
wines_cluster[0]
```

```python
c, coph_dists=cophenet(wines_cluster, pdist(wines_standardized))
```

```python
c
```

```python
coph_dists.shape
```

```python
plt.figure(figsize=(20, 16))
plt.title('Dendrogram')
plt.xlabel('Index Numbers')
plt.ylabel('Distance')
dendrogram(
    wines_cluster,
    leaf_rotation=90,
    leaf_font_size=8.,
    
)
# plt.ylim((0, 2))

plt.show()
```

```python
quality.value_counts()
```

There are 6 type of quality

We try to find 6 clusters from image which is around 39

```python
max_dist=39
clusters=fcluster(wines_cluster, max_dist, criterion='distance')
print set(clusters)
clusters
```
