```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('../datasets/votes.csv', index_col=0)
df
```

```python
y=df.Class
```

```python
for col in df.columns:
    df[col]=df[col].apply(lambda x: 1 if x=='y' else 0)
```

```python
x=df[[col for col in df.columns if col != 'Class']]
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
napolean=StandardScaler().fit_transform(x)
napolean.shape
```

```python
covariance_matrix=np.cov(napolean.T)
covariance_matrix.shape
```

```python
eignvalues, eigenvectors=np.linalg.eig(covariance_matrix)
```

```python
eignvalues
```

```python
eigenvectors.shape
```

```python
eigenpairs=zip(eignvalues, eigenvectors)
```

```python
eigenpairs
```

```python
for i in sorted(eigenpairs, reverse=True):
    print i
```

```python
eignsums=sum(eignvalues)
eignsums
```

```python
for eig in sorted(eigenpairs, reverse=True):
    print eig[0]/eignsums
```

```python
np.cumsum([(eig[0]/eignsums)*100.0 for eig in sorted(eigenpairs, reverse=True)])
```

```python
fit_me=[]
for eignpair in sorted(eigenpairs, reverse=True)[0:8]:
    fit_me.append(eignpair[1].reshape(16,1))
```

```python
# fit_me= [(eignpair[1].reshape(16,1) for eig in sorted(eigenpairs, reverse=True)[0:8])]
```

```python
fit_me
```

```python
Pmatrix=np.hstack(fit_me)
```

```python
pca_data=napolean.dot(Pmatrix)
```
