```python
%matplotlib inline

# Download the data, save to a file called "housing.data"
import urllib
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
urllib.urlretrieve (data_url, "housing.data")
```

```python
names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
         "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
```

```python
import pandas as pd

data = pd.read_csv("housing.data", header=None, names=names, delim_whitespace=True)

# Take a look at the first few rows
data.head()
```

```python
import seaborn as sns

# Plot using pandas
data.plot.scatter(x='CRIM', y='MEDV')
```

```python
# Knowledge Check Solution
from matplotlib import pyplot as plt

# Plot using matplotlib
plt.scatter(data["CRIM"], data["MEDV"])
plt.xlabel("Crime")
plt.ylabel("House Value")
```

```python
plt.scatter(data["ZN"], data["MEDV"])
plt.xlabel("proportion of residential land zoned for lots over 25,000 sq.ft")
plt.ylabel("House Value")
```

```python
plt.scatter(data["INDUS"], data["MEDV"])
plt.xlabel("proportion of non-retail business acres per town")
plt.ylabel("House Value")
```

```python
plt.scatter(data["NOX"], data["MEDV"])
plt.xlabel("nitric oxides concentration (parts per 10 million)")
plt.ylabel("House Value")
```

```python
plt.scatter(data["RM"], data["MEDV"])
plt.xlabel("average number of rooms per dwelling")
plt.ylabel("House Value")
```

```python
plt.scatter(data["AGE"], data["MEDV"])
plt.xlabel("proportion of owner-occupied units built prior to 1940")
plt.ylabel("House Value")
```

```python
plt.scatter(data["DIS"], data["MEDV"])
plt.xlabel("weighted distances to five Boston employment centres")
plt.ylabel("House Value")
```

```python
plt.scatter(data["RAD"], data["MEDV"])
plt.xlabel("index of accessibility to radial highways")
plt.ylabel("House Value")
```

```python
plt.scatter(data["TAX"], data["MEDV"])
plt.xlabel("full-value property-tax rate per $10,000")
plt.ylabel("House Value")
```

```python
plt.scatter(data["PTRATIO"], data["MEDV"])
plt.xlabel("pupil-teacher ratio by town")
plt.ylabel("House Value")
```

```python
plt.scatter(data["B"], data["MEDV"])
plt.xlabel("1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town")
plt.ylabel("House Value")
```

```python
plt.scatter(data["LSTAT"], data["MEDV"])
plt.xlabel("% lower status of the population")
plt.ylabel("House Value")
```
