```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/h02.csv', names=['period', 'sales'], index_col=0)
df.index = pd.to_datetime(df.index)
df.head()
```

```python
%matplotlib inline
```

```python
df.sales.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomp = seasonal_decompose(df.sales)
plot = decomp.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot = plot_acf(df.sales, lags=30)
```

```python
diff = df.sales.diff(periods=1)[1:]
diff.plot()
```

```python
for x in range(1, 25):
    print diff.autocorr(x)
```

```python
plot = plot_acf(diff, lags=30)
```

```python
import statsmodels.api as sm
```

```python
model = sm.tsa.statespace.SARIMAX(df, order=(1,1,1), 
                                  seasonal_order=(1,1,1,12))
results = model.fit()
results.summary()
```

```python
residuals = results.resid[1:]
plt = residuals.plot()
```
