```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
```

```python
names = [
    'id',
    'Clump Thickness',
    'Uniformity of Cell Size',
    'Uniformity of Cell Shape',
    'Marginal Adhesion',
    'Single Epithelial Cell Size',
    'Bare Nuclei',
    'Bland Chromatin',
    'Normal Nucleoli',
    'Mitoses',
    'Class'
]
```

```python
data = pd.read_csv('../datasets/breast-cancer-wisconsin.csv', names=names)
data
```

```python
data['Class'] = data['Class'].apply(lambda x: 1 if x == 4 else 0)
data['Class'].describe()
```

```python
sns.pairplot(data)
```

```python
data.corr()
```

```python
data.corr().loc[:, 'Class']
```

```python
y = data['Class'].reshape(-1, 1)
X = data[['Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Bland Chromatin']]
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X, y)
```

```python
knn.score(X, y)
```

```python
new_point = [[5, 8, 2]]
print knn.predict(new_point)
knn.predict_proba(new_point)
```
