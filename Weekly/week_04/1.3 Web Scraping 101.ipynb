```python
soup = BeautifulSoup(urllib2.urlopen('http://www.jcmit.com/memoryprice.htm').read())
table = soup.find('table')
headers = ['date (X)','$/Mbyte (Y)', 'Date', 'Month', 'Ref:', 'Page', 'Company', 'Size', 'Cost', 'Speed', 'Memory Type', 'Size Kbit', 'US$', 'nsec']
d = []
for row in table.find_all('tr')[2:]:
    d.append([str(td.text) for td in row.find_all('td')])
df = pd.DataFrame(d, columns=headers)
```
