```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import sqlite3
```

```python
sqlite_db = '../datasets/starter_code.sqlite'
conn = sqlite3.connect(sqlite_db)
c = conn.cursor()
```

```python
business = pd.read_csv("../datasets/businesses.csv")
inspections= pd.read_csv("../datasets/inspections.csv")
legend= pd.read_csv("../datasets/legend.csv")
violations= pd.read_csv("../datasets/violations.csv")
```

```python
create_businesses = "CREATE TABLE businesses (business_id varchar PRIMARY KEY, name varchar, address varchar, city varchar, state varchar, postal_code varchar, latitude varchar, longitude varchar, phone_number varchar);"
c.execute(create_businesses)
```

```python
create_inspections="CREATE TABLE inspections (business_id varchar, score varchar, date varchar, type varchar)"
c.execute(create_inspections)
```

```python
create_legend = "CREATE TABLE legend (Minimum_Score varchar, Maximum_score varchar, Descriptions varchar)"
c.execute(create_legend)
```

```python
create_violations = "CREATE TABLE violations (business_id varchar, date varchar, description varchar)"
c.execute(create_violations)
```
