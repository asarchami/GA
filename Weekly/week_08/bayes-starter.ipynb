```python
def square(x):
    return x**2

def weird_square(x):
    return x**2-2

def really_weird_square(x):
    return 2*(x**2)
```

```python
for x in [-5, -3, -1, 1, 3, 5]:
    print x, ' ', square(x), ' ', weird_square(x), ' ', really_weird_square(x)
```

```python
xs = [-5, -3, -1, 1, 3, 5]
squares=[square(x) for x in xs]
weird_squares=[weird_square(x) for x in xs]
really_weird_squares=[really_weird_square(x) for x in xs]
```

```python
plt.plot(xs, squares)
plt.plot(xs, weird_squares)
plt.plot(xs, really_weird_squares)
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(0,1.,.01)   # We want to ensure we have enough "granularity" in the graph so we set hash marks to 100
y = (x**5)*(1-x)**4  # Mapping the Bernoulli scheme

# Alter the prior and see how that visually alters the posterior

prior = (x**.46)*(1-x)**.32
z = y*prior

#plt.plot(x,prior)
plt.plot(x, y, 'r',x, z, 'g--')
plt.axvline(x=.55556,color='k',ls='dashed')

plt.show()
```
