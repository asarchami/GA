```python
import csv
import numpy as np
import scipy.stats as stats
import pylab as pl
```

```python
sat_csv_path = 'sat_scores.csv'
```

```python
list_data = []
with open(sat_csv_path, 'r') as f:
    for row in csv.reader(f):
        list_data.append(row)
# with open(sales_csv_path, 'r') as f:
#   sales_data = [x.split() for x in f]
# print list_data

# print list_data
labels = list_data[0]
del list_data[0]
list_data = np.array(list_data)
states = list_data[:,0]
# print states

state_data = {}
for i in range(0, len(list_data)):
    state_data[list_data[i][0]] = list_data[i]

    dic_data = {}
dic_data[labels[0]] = list_data[:, 0]
standard_deviations = {}
for i in range(1, len(labels)):
    dic_data[labels[i]] = list_data[:, i].astype(int)
    standard_deviations[labels[i]] = np.std(dic_data[labels[i]])
standard_deviations
```

```python
def plot_distribution(dataset):
    fit = stats.norm.pdf(dataset, np.mean(dataset), np.std(dataset)) 
    pl.hist(dataset,normed=True)
    pl.show()

plot_distribution(dic_data['Rate'])
```
